# Simple Linear Regression

```python
# ● Use the scikit-learn library. 
# ● Load the California Housing Dataset (or any house price dataset). 
# ● Predict the house price using a single feature (e.g., number of rooms). 
# ● Train a simple linear regression model. 
# ● Plot the regression line using matplotlib. 
```

```python
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.datasets import fetch_california_housing
```

```python
data = fetch_california_housing()
x = data.data 
y = data['target']
print(data.feature_names)
```

```python
# as AveRooms has 3 index in every row means 
# humhe saare rows chahiye pr unn rows mei se 2nd value so slicing
x1 = x[: , [2]]
x1
# here x1 is a single feature containing 2d-array of aveRooms
```

```python
# training the modelll

X_train, X_test, y_train, y_test = train_test_split(x1, y, test_size=0.2, random_state=42)

model = LinearRegression()
model.fit(X_train,y_train)

y_pred = model.predict(X_test)

print(y_pred)
print(y_test)

# difference is too much means error is high 
# model has less accuracy b'coz of single feature is used in training
```

```python
score = model.score(X_test, y_test)
score
```

```python
plt.scatter(x1,y)
# solo plot of correct values
```

```python
# display
plt.figure(figsize=(8,4))
#  s  is marker size
plt.scatter(X_test,y_test ,color='red',label='actual_price' , s=8)
plt.scatter(X_test,y_pred ,  color='green',label='Predicted_price' , s=8)
plt.xlabel('Average Rooms')
plt.ylabel('House Price')
plt.xlim(0,40)
plt.ylim(0,6)
plt.legend()
plt.grid()
plt.title("House Price Prediction")
```
